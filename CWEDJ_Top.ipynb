```python
import pandas as pd
```

```python
#loading top 50 original data and converting to data frame
top_25 = "/Users/grizzlyjenkins/Personal Documents/top_25_1.csv"
top_25_df = pd.read_csv(top_25)
top_25_df.sort_values('Rank', ascending=False)
top_25_df.head()
```

```python
#loading Amy's top 25 design research and converting to data frame
top_25_design = "/Users/grizzlyjenkins/Personal Documents/top_25_d.csv"
top_25_design_df = pd.read_csv(top_25_design)
top_25_design_df.sort_values('Rank', ascending=False)
top_25_design_df.head()
```

```python
#mergin the previously loaded data frames
merged_data = pd.merge(top_25_df,top_25_design_df,on='A')
merged_data.head()
```

```python
#cleaning up column names
top_25_merged = merged_data.rename(columns={"FA #_x": "FA #",
                                   "FA Name_x": "FA Name"})
top_25_merged.head()
```

```python
#loading in physical locations report and converting to data frame
pl_report = "/Users/grizzlyjenkins/Personal Documents/pl_report.csv"
pl_df = pd.read_csv(pl_report)
pl_df.head()
```

```python
#cleaning up column names
pl_df_clean = pl_df.rename(columns={"Branch #": "A"})
pl_df_clean.head()
```

```python
#merging physical locations data frame with the previously merged and cleaned data frames (top 50 and Amy's top 25 design info)
final_merge = pd.merge(pl_df_clean,top_25_merged,on='A')
final_merge
```

```python
final_merge.to_csv('/Users/grizzlyjenkins/Personal Documents/top_25_clean.csv')
```
